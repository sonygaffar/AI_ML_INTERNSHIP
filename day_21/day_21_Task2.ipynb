Machine Learning Types

1. Supervised Learning

Supervised Learning is when we give the computer data with the correct answers. 
The computer looks at the examples and learns how to predict the answers for new data.

Example: Email spam detection, predicting house prices.

```python
# Supervised Learning example

from sklearn.tree import DecisionTreeClassifier

# Data (numbers 1 to 6) and labels (0=odd, 1=even)
X = [[1], [2], [3], [4], [5], [6]]
y = [0, 1, 0, 1, 0, 1]

model = DecisionTreeClassifier()
model.fit(X, y)

# Predict new number
print("Prediction for 7:", model.predict([[7]])) 
```

2. Unsupervised Learning

Unsupervised Learning is when the computer is given data without any answers. 
It tries to find patterns or groups in the data on its own.

Example: Grouping customers by shopping habits, finding products often bought together.

```python
# Unsupervised Learning example
from sklearn.cluster import KMeans

# Data: numbers grouped in 2 clusters
X = [[1], [2], [3], [10], [11], [12]]

kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)

print("Cluster labels:", kmeans.labels_)
```

3. Reinforcement Learning

Reinforcement Learning is when the computer learns by trying things and getting rewards or penalties. 
It improves by learning what actions are good or bad.

Example: A robot learning to walk, a computer learning to play a game.

```python
# Reinforcement Learning example
actions = ["left", "right"]
correct_action = "right"
score = 0
for action in actions:
    if action == correct_action:
        score += 1  
    else:
        score -= 1  

print("Final score:", score)
```
