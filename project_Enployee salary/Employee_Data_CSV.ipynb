```python
import csv

# Data to write
data = [
    ["EmployeeID", "Name", "Department", "Experience", "Salary"],
    [1, "John", "IT", 5, 75000],
    [2, "Sarah", "HR", 3, 60000],
    [3, "Mike", "IT", 8, 120000],
    [4, "Anna", "Finance", 6, ""],   # Missing salary
    [5, "David", "HR", 10, 150000],
    [6, "Emily", "", 4, 70000],
    [7, "James", "IT", 2, 50000],
    [8, "Laura", "HR", 7, 90000],
    [9, "Robert", "Finance", 9, 110000],
    [10, "Linda", "IT", 1, 45000],
    [11, "Tom", "HR","", 160000],
    [12, "Sophia", "Finance", 3, 65000]
    
]

# Create CSV file
with open("employees.csv", mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("employees.csv file created successfully!")

import pandas as pd
import numpy as np
# Load dataset
df = pd.read_csv("employees.csv")

#Display first 5 rows
#print("First 5 Rows:")
print(df.head())



# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Fill missing salary with median salary
median_salary = df["Salary"].median()
df["Salary"].fillna(median_salary, inplace=True)

print("\nMissing Values After Handling:")
print(df.isnull().sum())

salaries = df["Salary"].values

print("\nSalary Statistics:")
print("Average Salary:", np.mean(salaries))
print("Median Salary:", np.median(salaries))
print("Minimum Salary:", np.min(salaries))
print("Maximum Salary:", np.max(salaries))
print("Standard Deviation:", np.std(salaries))

# Define high earners as salary > 100000 (example threshold)
high_earners = df[df["Salary"] > 100000]

print("\nHigh Earners:")
print(high_earners)

dept_analysis = df.groupby("Department")["Salary"].mean()

print("\nAverage Salary per Department:")
print(dept_analysis)

# Correlation between Salary and Experience
correlation = df["Salary"].corr(df["Experience"])

print("\nCorrelation between Salary and Experience:", correlation)

top_10_percent_threshold = np.percentile(df["Salary"], 90)
top_10_percent = df[df["Salary"] >= top_10_percent_threshold]

print("\nTop 10% Earners:")
print(top_10_percent)

```
