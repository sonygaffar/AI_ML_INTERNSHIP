```python
#Task 1: The Sample Space Map (Fundamentals)
import random
trials = 1000
count = 0
for i in range(trials):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    total = die1 + die2
    if total == 7:
        count = count + 1
probability = count / trials
print("Experimental probability of getting sum 7 is:", probability)
```
