```python
import pandas as pd

data = {
    "Age": ["23", "29", "25", "30"],
    "Salary": ["45000", "52000", "48000", "51000"],
    "Join_Date": ["2023-01-10", "2022-12-05", "2023-02-20", "2022-11-15"]
}

df = pd.DataFrame(data)
print(df.dtypes)
```

```python
#Safe Conversion with to_numeric()
df["Salary"] = pd.to_numeric(df["Salary"], errors="coerce")
print(df.dtypes)
```

```python
#Convert Dates (Critical for Analysis)
df["Join_Date"] = pd.to_datetime(df["Join_Date"])
print()
```

```python
import pandas as pd

data = {
    "Name": [" asha ", "RAHUL", "SiTa", "kiran "],
    "City": [" bangalore", "Mysuru ", "BANGALORE", "mysore"],
    "Gender": ["male", "Male ", " FEMALE", "female"],
    "Salary": ["₹45,000", "52000 ", " 48000", "₹51,000"]
}

df = pd.DataFrame(data)
print(df)
```

```python
#using strip to remove leading and trailing spaces
df["Name"] = df["Name"].str.strip()
df["City"] = df["City"].str.strip()
df["Gender"] = df["Gender"].str.strip()
df["Salary"] = df["Salary"].str.strip()
print(df)
```

```python
#using title to standardize text data
df["Name"] = df["Name"].str.title()
df["City"] = df["City"].str.title()
print(df)
```

```python
#convert to lowercase
df["Gender"] = df["Gender"].str.lower()
print(df)
```

```python
#replacing city names
df["City"] = df["City"].replace({"Mysore": "Mysuru"})
df["City"] = df["City"].replace({"Bangalore": "Bengaluru"})
print(df)
```

```python
#replacing special characters in Salary column using regex
df["Salary"] = df["Salary"].str.replace("₹", "")
df["Salary"] = df["Salary"].str.replace(",", "")
print(df)
```

```python
#replacing special characters in Salary column using regex
df["Salary"] = df["Salary"].str.replace("₹|,", "", regex=True)
print(df)
```

```python
#using astype to convert Salary to numeric
df["Salary"] = df["Salary"].astype(float)
print(df)
```

```python
# using unique to find unique values in City column
print(df["City"].unique())
```
