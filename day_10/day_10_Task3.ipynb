```python
#Task 3: The Categorical Standardizer (String Cleaning)
import pandas as pd
data = {
    "Location": [" New York", "new york", "NEW YORK ", "Los Angeles", " los angeles "]
}
df = pd.DataFrame(data)
print("Original Data:")
print(df)
df["Location"] = df["Location"].str.strip()
df["Location"] = df["Location"].str.title()
print("\nUnique Locations After Cleaning:")
print(df["Location"].unique())
print("\nCleaned Data:")
print(df)
```
