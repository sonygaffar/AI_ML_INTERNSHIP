```python
#Task 1: The Normalizer (Broadcasting & Stats)
import numpy as np
np.random.seed(0)
scores = np.random.randint(50, 101, size=(5, 3))
subject_means = scores.mean(axis=0)
centered_scores = scores - subject_means
print("Original Scores:")
print(scores)
print("\nSubject-wise Mean:")
print(subject_means)
print("\nCentered Scores:")
print(centered_scores)
```

```python
import numpy as np
scores = np.array([
    [60, 70, 80],
    [65, 75, 85],
    [70, 80, 90],
    [75, 85, 95],
    [80, 90, 100]
])

means = scores.mean(axis=0)
centered = scores - means
print("Original Scores:")
print(scores)
print("Column-wise Means:")
print(means)
print("Centered Scores:")
print(centered)
```
