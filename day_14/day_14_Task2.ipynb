```python
#Task 2: The Leveling Field (Feature Scaling)
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
df = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45, 50],
    'Salary': [30000, 50000, 70000, 90000, 110000, 130000]
})
standardized = StandardScaler().fit_transform(df)
normalized = MinMaxScaler().fit_transform(df)
plt.figure()
plt.hist(df['Salary'])
plt.title("Before Scaling")
plt.show()
plt.figure()
plt.hist(standardized[:, 1])  
plt.title("After Standardization")
plt.show()
plt.figure()
plt.hist(normalized[:, 1])  
plt.title("After Normalization")
plt.show()
```
