```python
# Task 3: The Complexity Creator (Polynomial Features)
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
X = np.arange(-5, 6).reshape(-1, 1)
y = X.flatten()**2
linear = LinearRegression().fit(X, y)
r2_linear = r2_score(y, linear.predict(X))
X_poly = PolynomialFeatures(2).fit_transform(X)
poly = LinearRegression().fit(X_poly, y)
r2_poly = r2_score(y, poly.predict(X_poly))
print("R² (Linear):", r2_linear)
print("R² (Polynomial):", r2_poly)
```
