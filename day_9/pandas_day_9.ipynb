```python
#Student attendance Example performing all the operations with name and who has more attendance and who has less attendance with attendance percentage and mark students are present or absent
import pandas as pd
names = ["Sony", "Bushra", "Akash", "Poornima", "Manasa", "Veda", "Sahil"]
attendance = pd.Series([10, 15, 20, np.nan, 25, 30, 35], index=names)
print("Original Attendance Series:")
print(attendance)

# Statistical operations
print("\nStatistical Operations:")
print("Mean:", attendance.mean())
print("Median:", attendance.median())
print("Mode:", attendance.mode())
print("Standard Deviation:", attendance.std())
print("Minimum Value:", attendance.min())
print("Maximum Value:", attendance.max())
print("Sum:", attendance.sum())
print("Count of Non-NaN Values:", attendance.count())

# Handling missing values
attendance_filled = attendance.fillna(0)
print("\nAttendance Series after filling NaN with 0:")
print(attendance_filled)

attendance_dropped = attendance.dropna()
print("\nAttendance Series after dropping NaN values:")
print(attendance_dropped)
```

```python
#Student attendance Example performing all the operations with name and who has more attendance and who has less attendance with attendance percentage and mark students are present or absent
import pandas as pd
names = ["Sony", "Bushra", "Akash", "Poornima", "Manasa", "Veda", "Sahil"]
attendance = pd.Series([35, 30, 10, np.nan, 25, 15, 34], index=names)
total_classes = 40
print("Original Attendance Series:")
print(attendance)

# Calculate attendance percentage
attendance_percentage = (attendance / total_classes) * 100
print("\nAttendance Percentage:")
print(attendance_percentage)

# Mark students as present or absent based on attendance percentage
attendance_status = (attendance_percentage >= 75).map({True: "Present", False: "Absent"})
print("\nAttendance Status:")
print(attendance_status)

# Statistical operations
print("\nStatistical Operations:")
print("Mean:", attendance.mean())
print("Median:", attendance.median())
print("Standard Deviation:", attendance.std())
print("Minimum Value:", attendance.min())
print("Maximum Value:", attendance.max())
print("Sum:", attendance.sum())
print("Count of Non-NaN Values:", attendance.count())

# Handling missing values
attendance_filled = attendance.fillna(0)
print("\nAttendance Series after filling NaN with 0:")
print(attendance_filled)
attendance_dropped = attendance.dropna()
print("\nAttendance Series after dropping NaN values:")
print(attendance_dropped)



```
