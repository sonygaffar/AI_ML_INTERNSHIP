```python
#Task 1: The Insight Filter (WHERE & GROUP BY)
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()
cursor.execute("""
CREATE TABLE interns (
    intern_id INTEGER PRIMARY KEY,
    intern_name TEXT,
    track TEXT,
    stipend INTEGER
)
""")
intern_data = [
    (1, "Sony", "Data Science", 6000),
    (2, "Bushra", "Web Development", 4500),
    (3, "Sahil", "Data Science", 5500),
    (4, "Akah", "Cybersecurity", 7000),
    (5, "Poornima", "Web Development", 5200),
    (6, "Manasa", "Data Science", 4800)
]
cursor.executemany("INSERT INTO interns VALUES (?, ?, ?, ?)", intern_data)
conn.commit()
filter_query = """
SELECT * 
FROM interns
WHERE track = 'Data Science' AND stipend > 5000
"""
filtered_df = pd.read_sql_query(filter_query, conn)
avg_query = """
SELECT track, AVG(stipend) AS average_stipend
FROM interns
GROUP BY track
"""
avg_df = pd.read_sql_query(avg_query, conn)
count_query = """
SELECT track, COUNT(*) AS intern_count
FROM interns
GROUP BY track
"""
count_df = pd.read_sql_query(count_query, conn)
print("Filtered Interns (Data Science, stipend > 5000):")
print(filtered_df)
print("\nAverage Stipend Per Track:")
print(avg_df)
print("\nIntern Count Per Track:")
print(count_df)
conn.close()
```
