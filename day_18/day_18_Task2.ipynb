```python
#Task 2: The Data Connector (JOINs & Python Integration)
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()
cursor.execute("""
CREATE TABLE interns (
    intern_id INTEGER PRIMARY KEY,
    intern_name TEXT,
    track TEXT
)
""")
cursor.execute("""
CREATE TABLE mentors (
    mentor_id INTEGER PRIMARY KEY,
    mentor_name TEXT,
    track TEXT
)
""")
intern_data = [
    (1, "Sony", "Data Science"),
    (2, "Bushra", "Web Development"),
    (3, "Manasa", "Data Science"),
    (4, "Veda", "Cybersecurity")
]
mentor_data = [
    (1, "Dr. Sahil", "Data Science"),
    (2, "Ms. Soni", "Web Development"),
    (3, "Mr. Dilbar", "Cybersecurity")
]
cursor.executemany("INSERT INTO interns VALUES (?, ?, ?)", intern_data)
cursor.executemany("INSERT INTO mentors VALUES (?, ?, ?)", mentor_data)
conn.commit()
query = """
SELECT 
    interns.intern_id,
    interns.intern_name,
    interns.track,
    mentors.mentor_name
FROM interns
INNER JOIN mentors
ON interns.track = mentors.track
"""
df = pd.read_sql_query(query, conn)
print(df)
conn.close()
```
