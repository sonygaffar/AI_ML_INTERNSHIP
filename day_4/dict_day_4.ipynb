Sets

```python
#create sets
s={1,2,3,4,5}
print(s)
#add element to set
s.add(6)
print(s)
#remove elements from det
s.remove(3)
print(s)
#check if element is in set
print(4 in s)
```

```python
#set union 
s1={4,5,6,7,3,4,2,1,5,4,5,6,7}
print(s1.union(s1))
#set intersection
print(s1.intersection(s1))
#set differece
print(s1.difference(s1))
#set symmetric difference
print(s1.symmetric_difference(s1))
```

```python
contacts = {
    "sony": "444-9000",
    "bushra": "567-3421",
    "manasa": "7894-2443"
}

contacts["sony"] = "555-4567"   
contacts["manasa"] = "555-9999"     

existing_lookup = contacts.get("sony", "Contact not found")
missing_lookup = contacts.get("veda", "Contact not found")

for name, phone in contacts.items():
    print(f"Contact: {name} | Phone: {phone}")

print("\nLookup existing:", existing_lookup)
print("Lookup missing:", missing_lookup)
```

```python
raw_logs = ["ID01", "ID02", "ID01", "ID05", "ID02", "ID08", "ID01"]
unique_users = set(raw_logs)
print("Unique users:", unique_users)
print('"ID05" is unique users ?', "ID05" in unique_users)
print("Original count:", len(raw_logs))
print("Unique count:", len(unique_users))
```

```python
friend_a= {"Python", "Cooking", "Hiking", "Movies"}
friend_b = {"Hiking", "Gaming", "Photography", "Python"}
ccc= friend_a & friend_b
bbb= friend_a | friend_b
eee = friend_a - friend_b
print("Shared interests:", ccc)
print("All interests:", bbb)
print("Unique to friend_a:", eee)
```
