```python
# plots.py
import matplotlib.pyplot as plt
import numpy as np

def generate_sample_plot():
    """Generate a simple line plot for testing purposes."""
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.figure()
    plt.plot(x, y)
    plt.title("Experimental Sine Wave Visualization")
    plt.xlabel("X values")
    plt.ylabel("sin(X)")
    plt.grid(True)
    plt.show()
if __name__ == "__main__":
    generate_sample_plot()
```
