```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Step 1: Load training data
df = pd.read_csv("homeprices.csv")
df
```

```python
# Step 2: Prepare features (X) and target (y)
X = df[['area']]
y = df['price']

print(X, y)
```

```python
# Step 3: Create and train model
model = LinearRegression()
model.fit(X, y)
print(model)
```

```python
# Step 4: Load test data
df_test = pd.read_csv("areas.csv")
print(df_test)
```

```python
# Step 5: Predict prices
df_test['price'] = model.predict(df_test)
print(df_test)
```

```python

# Step 6: Save predictions to new file
df_test.to_csv("predicted_prices.csv", index=False)
print("Predictions saved to predicted_prices.csv")
```

```python

# Optional: Print model equation
print("Coefficient (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
print("\nEquation: price =", model.coef_[0], "* area +", model.intercept_)
```

```python

# Optional: Plot regression line
#plt.scatter(df['area'], df['price'])
plt.plot(df['area'], df['price'])
plt.plot(df['area'], model.predict(X))
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```
