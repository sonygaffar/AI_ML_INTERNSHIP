```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#study hours
x=np.array([5,6,7,8,9])

#scores
y=np.array([40,50,45,85,89])
```

```python
x_mean=np.mean(x)
```

```python
y_mean=np.mean(y)
```

```python
x_mean , y_mean
```

```python
#calculate the slope (m)
numerator=np.sum((x-x_mean)*(y-y_mean))
denominator=np.sum((x-x_mean)**2)
m=numerator/denominator
```

```python
m
```

```python
#calculating b

b=y_mean-(m*x_mean)
b
```

```python
#predicting y for x=9
y_pred=m*x+b
y_pred
```

```python
#plt.scatter(x,y)
plt.plot(x,y)
plt.plot(x,y_pred)
plt.xlabel('Study Hours')
plt.ylabel('Scores')
plt.show()
```

```python
#using python libraries
from sklearn.linear_model import LinearRegression
x=x.reshape(-1,1)
model=LinearRegression()
model.fit(x,y) #trainig the model to find the best fit line
y_pred=model.predict(x)
y_pred
plt.plot(x,y)
plt.plot(x,y_pred)
plt.xlabel('Study Hours')
plt.ylabel('Scores')
plt.show()
```
