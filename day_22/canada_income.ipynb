```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df = pd.read_csv("canada_per_capita_income.csv")

X = df[['year']]
y = df['per capita income (US$)']

model = LinearRegression()
model.fit(X, y)

future_years = pd.DataFrame({
    'year': np.arange(2020, 2031)
})


future_years['predicted_income'] = model.predict(future_years)

future_years.to_csv("predicted_per_capita_income_2020_2030.csv", index=False)

print("Predicted Data:")
print(future_years)

plt.scatter(df['year'], df['per capita income (US$)'])
plt.plot(df['year'], model.predict(X))
plt.plot(future_years['year'], future_years['predicted_income'])
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.show()
```
