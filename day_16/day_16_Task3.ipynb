```python
#Task 3: The Magic of Averages (Central Limit Theorem)
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
data = np.random.lognormal(mean=10, sigma=1, size=10000)  
sample_means = []
for _ in range(1000):
    sample = np.random.choice(data, size=30, replace=True)
    sample_means.append(np.mean(sample))
plt.figure(figsize=(10,6))
plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Sample Means (n=30, 1000 samples)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
