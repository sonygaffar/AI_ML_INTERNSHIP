```python
#Task 1: The Shape Shifter (Visualizing Distributions)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
np.random.seed(10)
heights = np.random.normal(170, 10, 1000)
print("Normal Distribution (Heights)")
print("Mean:", np.mean(heights))
print("Median:", np.median(heights))
print()
plt.figure()
plt.hist(heights, bins=30, density=True)
kde = gaussian_kde(heights)
x = np.linspace(min(heights), max(heights), 1000)
plt.plot(x, kde(x))
plt.title("Normal Distribution - Heights")
plt.show()
income = np.random.exponential(50000, 1000)
print("Right Skewed Distribution (Income)")
print("Mean:", np.mean(income))
print("Median:", np.median(income))
print()
plt.figure()
plt.hist(income, bins=30, density=True)
kde = gaussian_kde(income)
x = np.linspace(min(income), max(income), 1000)
plt.plot(x, kde(x))
plt.title("Right Skewed Distribution - Income")
plt.show()
scores = 100 - np.random.exponential(15, 1000)
print("Left Skewed Distribution (Scores)")
print("Mean:", np.mean(scores))
print("Median:", np.median(scores))
print()
plt.figure()
plt.hist(scores, bins=30, density=True)
kde = gaussian_kde(scores)
x = np.linspace(min(scores), max(scores), 1000)
plt.plot(x, kde(x))
plt.title("Left Skewed Distribution - Scores")
plt.show()
```
