```python
#Task 2: The Outlier Detective (Z-Scores)
import numpy as np
import pandas as pd
data = {
    "student_name": ["A", "B", "C", "D", "E", "F", "G"],
    "score": [50, 55, 60, 58, 62, 59, 120]   # 120 is extreme value
}
df = pd.DataFrame(data)
mean = df["score"].mean()
std = df["score"].std()
print("Mean:", mean)
print("Standard Deviation:", std)
print()
df["z_score"] = (df["score"] - mean) / std
outliers = df[abs(df["z_score"]) > 3]
print("Full Dataset with Z-Scores:\n")
print(df)
print()
print("Outliers (|Z| > 3):\n")
print(outliers)
```
